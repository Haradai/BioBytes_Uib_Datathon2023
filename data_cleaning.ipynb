ANothetr approach, look at the big % values, what is the main thing living here?

```python
threshold = 1
sampls = df.columns[2:]

main_species_bysampl = {}
for s in sampls:
    main_species_bysampl[s] = np.where(df[s].to_numpy() > 1)
    
```

```python
# Concatenate all arrays
all_values = np.concatenate([arr for arr_tuple in main_species_bysampl.values() for arr in arr_tuple])

# Find unique values
unique_values = np.unique(all_values)

print(unique_values)
```

What are the main species we have?

```python
#get the labels for them
labels = df["Scientific Name"][unique_values]
print("Main species: ")
print(list(labels))
```

What are the main taxonomies we have?

```python
#get the labels for them
labels = df["Taxonomic Rank"][unique_values]
print("Main species: ")
print(list(labels))
```

```python
import re
import matplotlib.pyplot as plt
letters = {}
for l in labels:
    result_string = re.sub(r'\d+', '', l)
    if result_string not in letters.keys():
        letters[result_string] = 0
    else:
        letters[result_string] += 1

lbs = list(letters.keys())
plt.hist(letters.values())
plt.xticks(np.arange(len(lbs)),lbs)
```

```python
import csv
# Specify the input and output file names
input_file = 'competition_dataset/abundance_table.csv'
output_file = 'competition_dataset/abundance_table2.csv'

# Function to remove leading spaces from the CSV rows
def remove_leading_spaces_from_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile, open(output_file, 'w', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        for row in reader:
            cleaned_row = [cell.lstrip() for cell in row]
            writer.writerow(cleaned_row)

# Remove leading spaces and save the cleaned CSV
remove_leading_spaces_from_csv(input_file, output_file)
print(f"Leading spaces removed from {input_file} and saved to {output_file}")
```
