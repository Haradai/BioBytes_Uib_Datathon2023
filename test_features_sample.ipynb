```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
species_df = pd.read_csv("sample_by_specie_timewise.csv")

# drop all
# species_df = species_df.drop(columns=["id","Specie","System","Replica","slope","var","std","mean"])

species_df = species_df.drop(columns=["id", "Replica"])

# map system to int
species_df["System"] = species_df["System"].map({"A": 1, "B": 2})

# map "Specie" to numbers
species_le = LabelEncoder()
species_df["Specie"] = species_le.fit_transform(species_df["Specie"])

species_df.head()
```
